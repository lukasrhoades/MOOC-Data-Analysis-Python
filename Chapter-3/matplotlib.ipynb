Simple figure

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = np.array([2, 5, 7, 4, 7, 0, 3, 1, 9, 2])
```

```python
plt.plot(a)                   # Plot the points in the array a
plt.title("My first figure")  # Add a title to the figure
plt.xlabel("My x-axis")       # Give a label to the x-axis
plt.ylabel("My y-axis")       # Give a label to the y-axis
plt.show()                    # Tell matplotlib to output the figure,
# not strictly required in notebooks (but a bit neater)
```

```python
fig, ax = plt.subplots(2, 2)
print(ax.shape)
ax[0,0].plot(np.arange(6))           # top left
ax[0,1].plot(np.arange(6, 0, -1))    # top right
ax[1,0].plot((-1)**np.arange(6))     # bottom left
ax[1,1].plot((-1)**np.arange(1, 7))  # bottom right
plt.show()
```

Exercise 3.10 (subfigures)

```python
import numpy as np
import matplotlib.pyplot as plt

def subfigures(a):
    """Creates figure with two subfigures, where left subfigure is
    a line plot and second subfigure is a scatter plot. Both use
    the first two columns of input array as x and y coordinates
    respectively and scatter uses third and fourth columns for
    color and size of points respectively
    """
    fig, ax = plt.subplots(nrows=1, ncols=2)

    # Left subfigure
    ax[0].plot(a[:,0], a[:,1])

    # Right subfigure
    ax[1].scatter(x=a[:,0], y=a[:,1], c=a[:,2], s=a[:,3])

    plt.show()

def main():
    a = np.random.randint(0, 10, (10, 4))
    subfigures(a)

if __name__ == "__main__":
    main()
```

Suggested solution:

```python
import numpy as np
import matplotlib.pyplot as plt
 
def subfigures(a):
    fig, ax = plt.subplots(1,2, figsize=(6,3))
    ax[0].plot(a[:,0], a[:,1])
    ax[1].scatter(a[:,0], a[:,1], c=a[:,2], s=a[:,3])
    plt.show()
 
def main():
    n = 10
    a = np.random.randint(0, 10, (n, 3))
    print(a)
    print(a.dtype)
    print(a.shape)
    a = np.concatenate([np.arange(n)[:, np.newaxis], a], axis=1)
    subfigures(a)
 
if __name__ == "__main__":
    main()
```
